El presente codigo es muy util para aproximar el valor de una integral definida por medio de Cuadraturas Gaussianas y polinomios de legendre.


```python
import numpy as np
import matplotlib.pyplot as plt
```

La primera funcion declarada como gaussxw utiliza un arreglo de numpy para crear los puntos de muestreo y los pesos a partir de N segmentos de la integral.

```python
def gaussxw(N):
    x,w= np.polynomial.legendre.leggauss(N)
    return x,w
```

Los puntos y pesos conseguidos en la funcion anterior estan obtenidos en el intervalo de $[-1,1]$ pero necesitamos los puntos ajustados para nuestro intervalo de integracion desde (a) hasta (b) para esto usamos la funcion gaussawab que hace el ajuste y retorna los puntos y pesos arreglados.

```python
def gaussxwab(a, b, x, w):
    return 0.5*(b-a)*x + 0.5*(b+a), 0.5*(b-a)*w
```

Esta es la funcion a la que le calculamos la integral

```python
def func(x):
    return np.sin(x**2)
```

Este argumento recibe los puntos y pesos ajustados y aplica la siguiente regla para retornar el valor aproximado de la integral:
\begin{align}
\int_a^b {\rm{d}}x f(x) \approx \sum_{k=1}^{N} w_k f(x_k).
\end{align}

```python
def sumatoria(x_k, w_k, func):
    return np.sum(w_k * func(x_k))

```

Esta celda contiene los intervalos de integracion y el valor de N segmentos con el que aproximar la intefral ,estos valores globales junto con la funcion a evaluar se pueden modificar para calcular cualquier integral requerida, ya por ultimo se imprimen las aproximaciones en dos intervalos diferentes para comparar la efectividad.

```python
a, b = 0, np.pi  
N_values = range(1, 21) 
results = [] 
for n in N_values:
    x, w = gaussxw(n)
    x_k, w_k = gaussxwab(a, b, x, w)
    results.append(sumatoria(x_k, w_k, func))


print(f"Resultado con N=5: {results[4]:.8f}")
print(f"Resultado con N=20: {results[-1]:.8f}")
```

Por ultimo usando la biblioteca de matplotlib para hacer la figura comparativa del numero de segmentos por el valor real de la integral para verificar en que N la aproximacion es mejor (normalmente son N pequenos).

```python
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(N_values, results, 'o-', color='#1f77b4', markersize=8, linewidth=2)
plt.xlabel("Nmero de puntos (N)", fontsize=12, labelpad=10)
plt.ylabel("Valor de la integral", fontsize=12, labelpad=10)
plt.title(r"Convergencia de $\int_0^\pi \sin(x^2)\,dx$",
         fontsize=14, pad=20)
plt.axhline(y=results[-1], color='red', linestyle='--', alpha=0.7,
           label=f'N=20: {results[-1]:.8f}')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(fontsize=10, framealpha=0.9)
plt.tight_layout()

```
